```python
#Importing data
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("tennis1.txt")

#Assuming the outcome is always 1
outcomes = np.ones(len(data))

#Getting the bookmaker/expert predictions
#Count starts at 0 so 3 is actually the 4th column, etc.
expertsPredictions = np.zeros((len(data),4))

for indxi in range(len(data)):
    expertsPredictions[indxi,0] = data[indxi,3]
    expertsPredictions[indxi,1] = data[indxi,5]
    expertsPredictions[indxi,2] = data[indxi,7]
    expertsPredictions[indxi,3] = data[indxi,9]
        
```

```python
def AA_Brier(expertsPredictions,outcomes):
    
    #+1 here accounts for initial weights
    weights = np.zeros((len(expertsPredictions)+1,len(expertsPredictions[0]))) 
    E_loss = np.zeros((len(expertsPredictions),len(expertsPredictions[0]))) 
    gamma = np.zeros((len(expertsPredictions)))
    gamma_loss = np.zeros((len(expertsPredictions)))
    
    #initialize the first row with initial weights 1/N
    weights[0,0] = 1/len(expertsPredictions[0])
    weights[0,1] = 1/len(expertsPredictions[0])
    weights[0,2] = 1/len(expertsPredictions[0])
    weights[0,3] = 1/len(expertsPredictions[0])
    
    for indxi, i in enumerate(expertsPredictions):
        g1 = 0
        g0 = 0
        
        for indxj in range(len(i)):
        
            #normalising the weights
            totalWeights = sum(weights[indxi])
            temp = weights[indxi,indxj]/totalWeights
            
            #calculating gamma
            g0 = temp * np.exp(-2*((expertsPredictions[indxi][indxj])**2)) + g0
            g1 = temp * np.exp(-2*((1-expertsPredictions[indxi][indxj])**2)) + g1
            
            if indxj == (len(i)-1):
                g0 = -0.5 * np.log(g0)
                g1 = -0.5 * np.log(g1)
                gamma[indxi] = 0.5 - (g1-g0)/2
                
                #calculates the expert's losses and updates the weights
                for indxc in range(len(weights[0])):
                    E_loss[indxi,indxc] = (outcomes[indxi] - expertsPredictions[indxi,indxc])**2
                    weights[indxi+1,indxc] = weights[indxi,indxc]*np.exp(-2*E_loss[indxi,indxc])
                    
                #normalising the weights
                totalWeights = sum(weights[indxi+1])   
                weights[indxi+1] = weights[indxi+1]/totalWeights
            
                #Learner Loss
                gamma_loss[indxi] = (outcomes[indxi] - gamma[indxi])**2
    
    # creation of the graphs
    ''' 
    E_1 = np.zeros((len(expertsPredictions)))
    E_2 = np.zeros((len(expertsPredictions)))
    E_3 = np.zeros((len(expertsPredictions)))
    E_4 = np.zeros((len(expertsPredictions)))
    E_lossAVG = np.zeros((len(expertsPredictions)))
    
    for indxc in range(len(E_loss)):
        E_1[indxc] = E_loss[indxc,0] 
        E_2[indxc] = E_loss[indxc,1] 
        E_3[indxc] = E_loss[indxc,2] 
        E_4[indxc] = E_loss[indxc,3] 
        E_lossAVG[indxc] = (outcomes[indxc]-sum(expertsPredictions[indxc])*0.25)**2    
        
    time = np.arange(0,10087)
    plt.figure(0)
    plt.plot(time,np.cumsum(gamma_loss),label="L_l(t)")
    plt.plot(time,np.cumsum(E_1),label="E_1(t)")
    plt.plot(time,np.cumsum(E_2),label="E_2(t)")
    plt.plot(time,np.cumsum(E_3),label="E_3(t)")
    plt.plot(time,np.cumsum(E_4),label="E_4(t)")
    plt.ylabel("Loss")
    plt.xlabel("Time")
    plt.legend()
    
    plt.figure(1)
    plt.plot(time,(np.cumsum(E_1)-np.cumsum(gamma_loss)),label="E_1(t)-L_l(t)")
    plt.plot(time,(np.cumsum(E_2)-np.cumsum(gamma_loss)),label="E_2(t)-L_l(t)")
    plt.plot(time,(np.cumsum(E_3)-np.cumsum(gamma_loss)),label="E_3(t)-L_l(t)")
    plt.plot(time,(np.cumsum(E_4)-np.cumsum(gamma_loss)),label="E_4(t)-L_l(t)")
    plt.ylabel("Loss Differences")
    plt.xlabel("Time")
    plt.legend()
    
    plt.figure(2)
    plt.plot(time,(np.cumsum(E_1)-np.cumsum(E_lossAVG)),label="E_1(t)-L_avg(t)")
    plt.plot(time,(np.cumsum(E_2)-np.cumsum(E_lossAVG)),label="E_2(t)-L_avg(t)")
    plt.plot(time,(np.cumsum(E_3)-np.cumsum(E_lossAVG)),label="E_3(t)-L_avg(t)")
    plt.plot(time,(np.cumsum(E_4)-np.cumsum(E_lossAVG)),label="E_4(t)-L_avg(t)")
    plt.ylabel("Loss Differences")
    plt.xlabel("Time")
    plt.legend()'''
    
    return np.sum(gamma_loss)

AA_Brier(expertsPredictions,outcomes)
```
